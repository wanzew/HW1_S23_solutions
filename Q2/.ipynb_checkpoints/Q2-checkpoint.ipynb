Here we are going to solve some equality-constrained optimization problems with Newton's method. We are given a problem 

$$ \begin{align} \min_x \quad & f(x) \\ 
 \mbox{st} \quad & c(x) = 0 
 \end{align}$$
 
Which has the following KKT conditions for optimality:
$$\begin{align}
\nabla f(x) + \bigg[ \frac{\partial c}{\partial x}\bigg] ^T \lambda &= 0 \\ 
c(x) &= 0 
\end{align}$$

Since the gradient of a scalar-valued function (vector input, scalar output) is just the transposed jacobian when there is only 1 equality constraint, we can write this down using gradient notation:

$$\begin{align}
\nabla f(x) + \nabla c(x) \cdot \lambda &= 0 \\ 
c(x) &= 0 
\end{align}$$

Which is just a root-finding problem. To solve this, we are going to solve for a $z = [x^T,\lambda]^T$ that satisfies these KKT conditions.

```julia
let
    Q = [1.65539  2.89376; 2.89376  6.51521];
    q = [2;-3]
    cost(x) = 0.5*x'*Q*x + q'*x + exp(-1.3*x[1] + 0.3*x[2]^2) # cost function 
    contour(-1:.1:1,-1:.1:1, (x1,x2)-> cost([x1;x2]),title = "Cost Function",
            xlabel = "X₁", ylabel = "X₂",fill = true)
        plot!(-1:.1:1, -0.3*(-1:.1:1).^2 - 0.3*(-1:.1:1) .- .2,lw = 3,label = "constraint")
end
```

```julia
# @testset "2c" begin # POINTS = 1
let
    Q = [1.65539  2.89376; 2.89376  6.51521];
    q = [2;-3]
    
    # cost and constraint functions 
    cost(x) = 0.5*x'*Q*x + q'*x + exp(-1.3*x[1] + 0.3*x[2]^2)
    constraint(x) =  -0.3*x[1]^2 - 0.3*x[1] -.2 - x[2]
    
    function kkt_conditions(z)
        x = z[1:2]
        λ = z[3]
        
        # TODO: return the stationarity condition for the cost function f (∇f(x))
        [
            FD.gradient(cost, x) + FD.gradient(constraint, x)*λ;
            constraint(x)
        ]
    end
    
    # vectors for storing the x_k's and r(x_k)'s
    X = [zeros(3) for i = 1:10]
    R = [zeros(3) for i = 1:10]
    
    # the residual function we are driving to zero 
    r(x) = kkt_conditions(x)
    
    # initial point NOTE: Newton's method always needs an initial point.
    X[1] = [.0,-0.5,0]
    R[1] = r(X[1])
    
    # TODO: use Newton's Method to find x st r(x) = zeros(3)
    # You can exit your loop when the norm of the residual is < 1e-10
    
    # SOLUTION
    # main loop 
    for k = 1:9
        
        # TODO: use your newton_step function to take a step, then check convergence
        # store x values in X, and store the residuals in R 
        # hint: R[i] = r(X[i])
        
        # SOLUTION 
        
        # take the newton step 
        J = FD.jacobian(r,X[k])
        X[k+1] = X[k] - J\R[k]

        # evaluate the residual to check for convergence
        R[k+1] = r(X[k+1])
        @show norm(R[k+1])
        if norm(R[k+1])<1e-10
            X = X[1:k+1]
            R = R[1:k+1]
            break
        end
        
    end
    
    
    # plotting stuff
    Rp = [[abs(R[i][ii]) + 1e-15 for i = 1:length(R)] for ii = 1:length(R[1])] # this gets abs of each term at each iteration
    
    plot(Rp[1],yaxis=:log,ylabel = "|r|",xlabel = "iteration",
         yticks= [1.0*10.0^(-x) for x = float(15:-1:-2)],
         title = "Convergence of Newton's Method on KKT Conditions",label = "|r_1|")
    plot!(Rp[2],label = "|r_2|")
    display(plot!(Rp[3],label = "|r_3|"))
    
    contour(-1:.1:1,-1:.1:1, (x1,x2)-> cost([x1;x2]),title = "Cost Function",
            xlabel = "X₁", ylabel = "X₂",fill = true)
    x1_hist = [x[1] for x in X]
    x2_hist = [x[2] for x in X]
    plot!(x1_hist, x2_hist, marker = :d, label = "xₖ")#, seriestype=:scatter)
    display(plot!(-1:.1:1, -0.3*(-1:.1:1).^2 -0.3*(-1:.1:1) .- .2,lw = 3,label = "constraint"))

end
```

```julia
# function cost(model, x_guess, z)
#     x = z[1:30]
#     u = z[31:42]
    
#     return 0.5*
function qdot(q1,q2)
    v1 = q1[2:4]
    s1 = q1[1]
    v2 = q2[2:4]
    s2 = q2[1]
    return [(s1 * s2 - dot(v1, v2));s1 * v2 + s2 * v1 + cross(v1, v2)]
end
function q_from_mrp(p)
    (1/(1+dot(p,p)))*[(1-dot(p,p));2*p]
end
function dynamics_mrp(params::NamedTuple, x::AbstractVector{T1}, u::AbstractVector{T2}) where {T1,T2}
    model = params.model
    x_guess = params.x_guess
    T = promote_type(T1,T2)
    state = model.statecache[T]
    res = model.dyncache[T]
    
    x_updated = zeros(T1, params.nx)
    x_updated[1:4] = qdot(x_guess[1:4], q_from_mrp(x[1:3]))
    x_updated[5:end] = x[4:end]

    # Convert from state ordering to the ordering of the mechanism
    copyto!(state, x_updated)
    τ = [zeros(3); u]
    dynamics!(res, state, τ)
    q̇ = res.q̇
    v̇ = res.v̇
    return [q̇; v̇]
end
let
    model = UnitreeA1()
#     mvis = initialize_visualizer(model)
#     render(mvis)
    
    x_guess = initial_state(model)
    
    x_initial = initial_state(model)
    u_initial = zeros(control_dim(model))
        
    params = (
            model = model, 
            nx = state_dim(model), 
            nu = control_dim(model),
            x_guess = x_guess
        )
    
    @show length(x_guess)
    @show length(u_initial)
    
    @show typeof(x_guess)
    
end
```
